```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris_data = load_iris(as_frame=True)
data = iris_data.frame
```

```python
random_sample = data.sample(n=30, random_state=42)
print("\nRandom Sample (30 observations):")
print(random_sample)
```

```python
sample_means = []
for _ in range(100):
    sample = data.sample(n=30, random_state=np.random.randint(0, 1000))
    sample_means.append(sample['sepal length (cm)'].mean())
```

```python
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=15, edgecolor='k', alpha=0.7)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean of Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
def systematic_sampling(data, sampling_fraction):
    interval = int(1 / sampling_fraction)
    return data.iloc[::interval]

systematic_sample = systematic_sampling(data, 0.2)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data.index, data['sepal length (cm)'], label='Original Data', alpha=0.5)
plt.scatter(systematic_sample.index, systematic_sample['sepal length (cm)'], label='Systematic Sample', color='red', alpha=0.7)
plt.title("Original vs. Systematic Sample")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.legend()
plt.show()
```
